```python
from train_tokenizer import read_training_info
import os
from transformers import AutoTokenizer
from datasets import Dataset
import json

# new_tokenizer_path = "/cmlscratch/astein0/efficient_tokenization_for_inference/tokenizers/phi_gsm8k_cot-1000"
# base_tokenizer_path = "microsoft/phi-4-mini-reasoning"

new_tokenizer_path = "/cmlscratch/astein0/efficient_tokenization_for_inference/tokenizers/Llama-3.2-tokenizer-magpie_pro_300k_filtered-math-empty-start-1000"
base_tokenizer_path = "meta-llama/Llama-3.2-3B-Instruct"
info_file = "training_info.json"
new_tokenizer_info = read_training_info(os.path.join(new_tokenizer_path, info_file))
# get merges and new_tokens
merges = new_tokenizer_info["merges"]
new_tokens = new_tokenizer_info["new_tokens"]
sizes = new_tokenizer_info["sizes"]
original_size = new_tokenizer_info["static_info"]["original_dataset_size"]
tokenizer = AutoTokenizer.from_pretrained(new_tokenizer_path)
base_tokenizer = AutoTokenizer.from_pretrained(base_tokenizer_path)

tokenizer_json = json.loads(tokenizer._tokenizer.to_str())

merges_list = tokenizer_json["model"]["merges"]
vocab_list = tokenizer_json["model"]["vocab"]
vocab = tokenizer.get_vocab()
threshold = len(base_tokenizer)
```

```python
print(len(merges))
print(len(new_tokens))
counter = 0

denom = sum([x / original_size for x in sizes])
merge_table = {x: (y, z/(original_size*denom)) for x, y, z in zip(new_tokens, merges, sizes)}

def deconstruct_merge(merge_list, new_merge_table, vocab_map, threshold: int):
    output_merge_list = []
    stack = merge_list.copy()
    stack.reverse()
    while stack:
        this_merge = stack.pop()
        token_id = vocab_map[this_merge] # if not found theres a problem
        if token_id < threshold:
            output_merge_list.append(this_merge)
        else:
            (next1, next2), _ = new_merge_table[this_merge]  # if not found then it should have been below the threshold
            stack.append(next2)
            stack.append(next1)
    return output_merge_list

dict_samples = []

for i, (new_token, (_, weight)) in enumerate(merge_table.items()):
    deconstructed_merge = deconstruct_merge([new_token], merge_table, vocab, threshold)
    merge_str = [tokenizer.convert_tokens_to_string([x]) for x in deconstructed_merge]
    print(f"{i}: {new_token} -> {merge_str} ({weight})")
    new_token_str = tokenizer.convert_tokens_to_string([new_token])
    if new_token_str != ''.join(merge_str):
        merge_str = None
        counter += 1
    # assert new_token_str == ''.join(merge_str), f"{i}: New token {new_token_str} is merge {merge_str}"
    sample = {"old_tokens": merge_str, "new_token": new_token_str, "token_id": vocab[new_token], "merge_ids": [vocab[token] for token in deconstructed_merge], "weight": weight}
    dict_samples.append(sample)

print(counter)
print(dict_samples[0])
ds_dict = Dataset.from_list(dict_samples)
print(ds_dict[0])
ds_dict.save_to_disk("datasets/llama-3.2-1000_dict")
```
